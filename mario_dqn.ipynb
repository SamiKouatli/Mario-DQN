# Training
To train model from scratch, run the following command :

```
python <path-to-__main__.py> train
```

Futhermore, two boolean arguments *--record* and *--verbose* can help you understand the code.


````python
# Train from scratch
!python . train --verbose True
````

The model is saved periodically, to changed its value : see *config/runner.yaml* => *score_buffer_size* parameter.

The save path is in *./save/*<*timestamp*>*/*. This folder contains results, video recordings and saved weights.

# Inference
To infer a trained model :


```
python <path-to-__main__.py> infer --path <path-to-trained-weights.h5>
```

Exactly like the "*train*" action, you can set *--record* and *--verbose *arguments.

````python
# Infer model
!python . infer --path ./save/cpu_trained/weights.h5
````

# Continue training a model
The *--path* arguments is also available for "*train*" action, it allows you to continue training a trained model :


```
python <path-to-__main__.py> train -- path <path-to-trained-weights.h5>
```



````python
# Train the already trained model
!python . train --path ./save/cpu_trained/weights.h5
````

# Plot results

Results are averaged over several episodes and saved in a result file. See *score_buffer_size* parameter.

If you want to plot the evolution of results during training :

```
python <path-to-__main__.py> plot -- path <path-to-result.txt>
```

Results can decrease during training, it happens because behaviour which do not accumulate reward are punished. Hence, when Mario reaches an obstacle that he has never encountered before, he may get stuck temporarily and thus accumulate punishments.


````python
# Plot the evolution of results over training
!python . plot --path ./save/cpu_trained/result.txt
````
